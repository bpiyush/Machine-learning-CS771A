```python
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import import_ipynb
import ppca_altopt
import pickle
import numpy as np
```

```python
with open('data/mnist_small_py2_3.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
X = data['X']
X.shape
```

### kMeans on PCA transformed data 

```python
Z, W = ppca_altopt.alt_opt_ppca(X, 2)
```

```python
Z.shape
```

```python
kmeans = KMeans(n_clusters=10, random_state=0).fit(Z)
labels = kmeans.labels_
```

```python
import matplotlib.cm as cm
import matplotlib.pyplot as plt
def visualize(X, labels, title):
    plt.figure(figsize=(15,10))
    plt.title(title)
    plt.xlabel('Axis 1')
    plt.ylabel('Axis 2')
    plt.scatter(X[labels==i][:,1], X[labels==1][:,1])
    plt.show()
```

```python
visualize(Z, labels, 'PCA')
```
